# NumPy 배열
- numpy에서 배열은 **ndarray** 또는 **array**라고 부름
- numpy.array와 python.array는 **다름**
  - numpy array : numpy array끼리 연산 가능
  - python list : 덧셈만 가능

  - numpy array : array 전체에 연산이 가능
  - python list : 곱셈만 가능 (요소 반복)

```python
p_list = [1, 2, 3]
n_array = np.array([1,2,3])

print(p_list + p_list)
print(n_array + n_array)

print(p_list * 3)
print(n_array * 3)
```

## NumPy 배열 예제

```python
# 2개의 축(axis)을 가지며, 첫번째 축은 길이가 2이고, 두번째 축은 길이가 3인 배열
# == 2행 3열의 2차원 배열
[[1, 0, 0],
 [0, 1, 2]]
```

```python
# 1개의 축을 가지며, 축은 3가지 요소(element)를 가졌으며, 길이(length)는 3이다
[1, 2, 1]
```

# NumPy 배열 속성값
- ndarray.**shape** : 배열의 각 축(axis)의 크기
- ndarray.**ndim** : 축의 개수(dimention)
- ndarray.**dtype** : 각 요소(element)의 타입
- ndarray.**itemsize** : 각 요소의 타입의 bytes 크기
- ndarray.**size** : 전체 요소의 개수

## NumPy 배열 속성값 예제

```python
# 3 * 4크기의 2차원 배열 생성
a = np.arange(12).reshape(3, 4)
print(a)
```

```python
# 몇 행 * 몇 열의 배열인지
a.shape
```

```python
# 몇 차원 배열인지
a.ndim
```

```python
# 요소의 데이터타입
a.dtype
```

```python
# 요소의 데이터타입의 bytes 크기
a.itemsize
```

```python
# 전체 요소 개수
a.size
```
